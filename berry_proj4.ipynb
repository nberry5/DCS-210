```r
glimpse(Fingers)
```

```r
gf_histogram(~ Thumb, data = Fingers, fill = ~ Sex) %>%
 gf_refine(scale_color_manual(values = c("purple", "orange")))
```

```r
gf_point(Thumb ~ Sex, data = Fingers)
```

```r
gf_jitter(Thumb ~ Sex, data = Fingers)
```

```r
# The jitter plot adds random noise to the data, making it easier to visualize
# the spread of the data, rather than having all data only on two vertical
# lines. This is necessary because Sex is a discrete variable, rather than
# continuous.
```

```r
gf_jitter(Thumb ~ Sex, data = Fingers, alpha = 0.2)

# It adjusts the opacity and transparency of the plot points. Lower alpha values
# mean more transparency. It's on a scale from 0-1.
```

```r
gf_boxplot(Thumb ~ Sex, data = Fingers)
```

```r
gf_boxplot(Thumb ~ Sex, data = Fingers, color = "orange") %>%
gf_jitter(alpha = 0.4)
```

```r
gf_point(Thumb ~ Index, data = Fingers)
```

```r
gf_boxplot(Thumb ~ Index, data = Fingers, color = "orange") %>%
gf_point(alpha = 0.5)
```

```r
# I used index finger length to predict thumb length. This was a much more
# reliable predictor than sex. Though males tended to have longer thumbs than
# females, we see a direct and tight correlation when using index finger. This
# is also helped by the fact that index finger length is a continuous variable,
# which allows us to see a linear trend better than we could when predicting
# a continuous variable with a discrete variable.
```
