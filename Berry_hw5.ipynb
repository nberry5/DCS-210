```r
install.packages("tidyverse")
library(tidyverse)
```

```r
probability = c()

count_pairs = 0
for (x in 1:10000) {
  pair_roll = sample(1:6, 5, replace=T)
  results = c()
  pair = FALSE
  for (die in pair_roll) {
    if (die %in% results) {
      pair = TRUE
    }
    results <- c(results, die)
  }
  if (pair) {
    count_pairs = count_pairs + 1
  }
}

pair_probability = count_pairs / 10000

pair_prob2 = 5400/7776
probability = c(probability, pair_prob2)

three_prob = 1200/7776
probability = c(probability, three_prob)

small_straight = 960/7776
probability = c(probability, small_straight)

large_straight = 240/7776
probability = c(probability, large_straight)

full_house = 300/7776
probability = c(probability, full_house)

yahtzee_prob = 6/7776
probability = c(probability, yahtzee_prob)
print(probability)
```

```r
dice_roll = c(
  "Pair", 
  "Three of a Kind", 
  "Small Straight", 
  "Large Straight",
  "Full House",
  "Yahtzee"
)
```

```r
yahtzee_probabilities <- data.frame(dice_roll, probability)
```

```r
yahtzee_probabilities
```

```r
yahtzee <- ggplot(data=yahtzee_probabilities, aes(x=dice_roll, y=probability)) +
  geom_bar(stat="identity", color = "#000000", fill = "#77AAFF") +
  geom_text(aes(label=round(probability, digits = 3)), vjust=-0.3, size=3.5) +
  labs(title="Probabilities of Winning in Yahtzee When Rolling All Five Dice", 
         x="Winning Roll", y = "Probability") +
  theme_classic()

yahtzee

# It is much more likely to roll a pair than any other winning roll (in fact,
# a pair is even more likely than not to win at all). Rolling a Yahtzee with
# five dice at once is so improbable that it's about the same likelihood of
# rolling 123 small straights before a single Yahtzee.
```
