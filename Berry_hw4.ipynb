```r
brexit <- read.csv("/content/brexit.csv")
```

```r
glimpse(brexit)
```

```r
ggplot(brexit, aes(x = opinion, fill = opinion)) +
  geom_bar() +
  facet_wrap(~region, nrow = 1) +
  theme(
    axis.ticks.x=element_blank(),
    axis.text.x=element_blank()
  ) +
  labs(
    x = "Opinion",
    y = "Count"
  ) +
  scale_fill_manual(values = c(
    "Wrong" = "#ef8a62",
    "Right" = "#67a9cf",
    "Don't know" = "#BBBBBB"
  ))
```

```r
ggplot(brexit, aes(y = region, fill = opinion)) +
  geom_bar() +
  facet_wrap(~opinion, nrow = 1) +
  guides(fill = "none") +
  labs(
    x = "Count",
    y = "Region"
  ) +
  scale_fill_manual(values = c(
    "Wrong" = "#ef8a62",
    "Right" = "#67a9cf",
    "Don't know" = "#BBBBBB"
  ))
```

```r
lon_dk = 24/(24+63+110)
lon_r = 63/(24+63+110)
lon_w = 110/(24+63+110)

mid_dk = 57/(57+145+152)
mid_r = 145/(57+145+152)
mid_w = 152/(57+145+152)

nor_dk = 48/(48+176+176)
nor_r = 176/(48+176+176)
nor_w = 176/(48+176+176)

ros_dk = 49/(49+241+257)
ros_r = 241/(49+241+257)
ros_w = 257/(49+241+257)

sco_dk = 10/(10+39+92)
sco_r = 39/(10+39+92)
sco_w = 92/(10+39+92)
```

```r
std_region <- c(
  "London", "London", "London", 
  "Midlands-Wales", "Midlands-Wales", "Midlands-Wales", 
  "North", "North", "North", 
  "Rest of South", "Rest of South", "Rest of South", 
  "Scotland", "Scotland", "Scotland"
)

std_opinion <- c(
  "Don't Know", "Right", "Wrong",
  "Don't Know", "Right", "Wrong",
  "Don't Know", "Right", "Wrong",
  "Don't Know", "Right", "Wrong",
  "Don't Know", "Right", "Wrong"
)

std_count <- c(
  lon_dk, lon_r, lon_w,
  mid_dk, mid_r, mid_w,
  nor_dk, nor_r, nor_w,
  ros_dk, ros_r, ros_w,
  sco_dk, sco_r, sco_w
)
```

```r
std_brexit <- data.frame(std_region, std_opinion, std_count)
```

```r
std_brexit
```

```r
ggplot(data = std_brexit, aes(x = std_count, y = std_region, fill = std_opinion)) +
  geom_bar(stat="identity", position=position_dodge()) +
  labs(
    x = "Proportion of Region's Responses",
    y = "Region",
    fill = "Opinion"
  ) +
  scale_fill_manual(values = c(
    "Wrong" = "#ef8a62",
    "Right" = "#67a9cf",
    "Don't know" = "#BBBBBB"
  ))
```
